### LEVEL 2 -(Task 1) Table booking and online delivery

```python
# Import necessary libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
```

```python
# Loading the dataset
data = pd.read_csv("Dataset .csv")
data
```

```python
data.head(10)
```

```python
# Assuming 'data' is your DataFrame
# Step 1: Identify categorical columns and there is List of columns to encode
categorical_columns = [ 'Has Table booking','Has Online delivery','Is delivering now','Switch to order menu','Rating text'] 

# Step 2: Instantiate LabelEncoder
label_encoder = LabelEncoder()

# Step 3: Fit and Transform each column
for column in categorical_columns:
    data[column] = label_encoder.fit_transform(data[column])

# Display the updated DataFrame
data.head()
```

```python
true_booking_counts =  data['Has Table booking'].value_counts()[1]
true_booking_counts
```

```python
true_online_delivery_count = data['Has Online delivery'].value_counts()[1]
true_online_delivery_count
```

```python
Total_restaurants = len(data)
table_booking_percentage = ( true_booking_counts / Total_restaurants) * 100

# Calculate percentage of restaurants offering online delivery
online_delivery_percentage = (true_online_delivery_count / Total_restaurants) * 100

print(f"Percentage of restaurants offering table booking {table_booking_percentage:.2f}%")
print(f"Percentage of restaurants offering table booking {online_delivery_percentage:.2f}%")
```

```python
# restraurant having both online delivery and table booking.
both = (data['Has Table booking'].value_counts()[1] & data['Has Online delivery'].value_counts()[1])
both
```

```python
# Calculate the percentage
Total_restaurants = len(data)
percentage = ( both / Total_restaurants) * 100

print("Percentage of restaurants offering both table booking and online delivery:", percentage)
```

```python
 # Calculate the average rating for restaurants with table booking
avg_rating_with_table_booking =  data[data['Has Table booking'] == True]['Rating text'].mean()

# Calculate the average rating for restaurants without table booking
avg_rating_without_table_booking = data[data['Has Table booking'] == False]['Rating text'].mean()

print(f"Average rating of restaurants with table booking{avg_rating_with_table_booking:.2f}%")
print(f"Average rating of restaurants without table booking {avg_rating_without_table_booking:.2f}%")
```


# LEVEL 2 - (Task 2)

### Price Range Analysis

```python
Price_range_count =  data['Price range'].value_counts()
Price_range_count
```

```python
# Find the most common price range
most_common_price_range = data['Price range'].mode()[0]
print("common price range is:", most_common_price_range)
```

```python
# Calculate the average rating for each price range
average_rating_by_price_range =data.groupby('Price range')['Rating text'].mean()

print("Average rating for each price range:")
print(average_rating_by_price_range)
```

```python
# Group by Price Range and calculate average rating
avg_rating_by_price = data.groupby('Price range')['Rating text'].mean()

# Find price range with highest average rating
highest_avg_rating_price = avg_rating_by_price.idxmax()

# Get the color associated with the price range with the highest average rating
color_for_highest_avg_rating = data.loc[data['Price range'] == highest_avg_rating_price, 'Rating color'].iloc[0]

print("Price range with the highest average rating:", highest_avg_rating_price)
print("Color associated with the highest average rating:", color_for_highest_avg_rating)
```
