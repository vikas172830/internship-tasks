# Level 1-> Task 3: Geospatial Analysis

Visualizing the locations of restaurants on a
map using latitude and longitude
information.

```python
pip install folium
```

```python
# Initialize a map centered around a specific location (e.g., the average latitude and longitude)
import folium

center_latitude = data['Latitude'].mean()
center_longitude = data['Longitude'].mean()
restaurant_map = folium.Map(location=[center_latitude, center_longitude], zoom_start=12)
```

```python
restaurant_map
```

```python
# Add markers to the map
for index, row in data.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(restaurant_map)

# Save the map to an HTML file
restaurant_map.save('restaurant_map.html')
```
